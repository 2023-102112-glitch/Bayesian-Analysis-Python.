```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = "/content/StudentsPerformance.csv.csv"
df = pd.read_csv(path)
```

```python
subset = df.head(15).copy()
subset['student_id'] = [f"S{i}" for i in subset.index]
```

```python
G = nx.Graph()
```

```python
for _, row in subset.iterrows():
    student = row['student_id']
    race = row['race/ethnicity']
    edu = row['parental level of education']

    # Add student nodes and their attribute nodes
    G.add_node(student, type='student')
    G.add_node(race, type='attribute')
    G.add_node(edu, type='attribute')

    # Add edges to show the relationship
    G.add_edge(student, race)
    G.add_edge(student, edu)
```

```python
pos = nx.spring_layout(G, seed=42)
color_map = ['skyblue' if G.nodes[node].get('type') == 'student' else 'lightgreen' for node in G.nodes()]
```

```python
plt.figure(figsize=(10, 7))
nx.draw(G, pos, with_labels=True, node_color=color_map, node_size=1500, font_size=8, font_weight='bold')
plt.title("Network of Students and their Attributes")
```

```python
plt.savefig('students_network_graph.png')
print("Nodes in graph:", G.nodes())
print("Edges in graph:", G.edges())
```
